```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Updated forecast horizons (now includes 72h)
horizons = [12, 24, 48, 72]

# Updated model performance with 72-hour LSTM results
data = {
    'Horizon': horizons,
    'Linear Regression MAE': [696.04, 724.70, 732.64, 778.93],
    'Random Forest MAE': [586.70, 627.20, 649.33, 685.43],
    'LSTM MAE': [245.16, 361.16, 421.01, 554.02],
    'GRU MAE' : [152.60, 150.34, 161.57, 177.23],
    'STGCN MAE' : [278.45, 244.16, 245.16, 235.61],
    'CNN-GRU-LSTM MAE' : [133.50, 124.32, 150.00, 171.02],
    #'STGCN-GRU-LSTM MAE' : [218.341,]
    
    
    

    'Linear Regression RMSE': [1027.22, 1062.04, 1069.93, 1120.21],
    'Random Forest RMSE': [866.81, 925.22, 955.91, 995.60],
    'LSTM RMSE': [388.32, 574.77, 671.78, 846.77],
    'GRU RMSE' : [306.13, 297.03, 355.28, 357.13],
    'STGCN RMSE': [478.31, 435.87, 430.14, 426.59],
    'CNN-GRU-LSTM RMSE' : [253.95, 241.41, 280.81, 326.84],
    #"STGCN-GRU-LSTM RMSE' : []
    
    
    
}

df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt

# --- MAE Plot ---
plt.figure(figsize=(8, 5))

plt.plot(df['Horizon'], df['Linear Regression MAE'], marker='o', linewidth=2, label='Linear Regression')
plt.plot(df['Horizon'], df['Random Forest MAE'], marker='s', linewidth=2, label='Random Forest')
plt.plot(df['Horizon'], df['LSTM MAE'], marker='^', linewidth=2, label='LSTM')
plt.plot(df['Horizon'], df['GRU MAE'], marker='d', linewidth=2, label='GRU')
plt.plot(df['Horizon'], df['STGCN MAE'], marker='v', linewidth=2, label='STGCN')
plt.plot(df['Horizon'], df['CNN-GRU-LSTM MAE'], marker='P', linewidth=2, label='CNN-GRU-LSTM')
# plt.plot(df['Horizon'], df['STGCN-GRU-LSTM MAE'], marker='X', linewidth=2, label='STGCN-GRU-LSTM')

plt.title('MAE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
plt.xticks(horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(fontsize=9, loc='upper left', ncol=2)  # Compact legend
plt.tight_layout()
plt.savefig('mae_line_plot.pdf', bbox_inches='tight')
plt.show()
```

```python
# --- RMSE Plot ---
plt.figure(figsize=(8, 5))

plt.plot(df['Horizon'], df['Linear Regression RMSE'], marker='o', linewidth=2, label='Linear Regression')
plt.plot(df['Horizon'], df['Random Forest RMSE'], marker='s', linewidth=2, label='Random Forest')
plt.plot(df['Horizon'], df['LSTM RMSE'], marker='^', linewidth=2, label='LSTM')
plt.plot(df['Horizon'], df['GRU RMSE'], marker='d', linewidth=2, label='GRU')
plt.plot(df['Horizon'], df['STGCN RMSE'], marker='v', linewidth=2, label='STGCN')
plt.plot(df['Horizon'], df['CNN-GRU-LSTM RMSE'], marker='P', linewidth=2, label='CNN-GRU-LSTM')
# Uncomment when you add results
# plt.plot(df['Horizon'], df['STGCN-GRU-LSTM RMSE'], marker='X', linewidth=2, label='STGCN-GRU-LSTM')

plt.title('RMSE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Root Mean Squared Error (RMSE)', fontsize=12)
plt.xticks(horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(fontsize=9, loc='upper left', ncol=2)
plt.tight_layout()
plt.savefig('rmse_line_plot.pdf', bbox_inches='tight')
plt.show()
```

```python
bar_width = 0.12   # smaller width to fit more models
x = np.arange(len(horizons))

plt.figure(figsize=(10, 6))

plt.bar(x - 3*bar_width, df['Linear Regression MAE'], width=bar_width, label='Linear Regression')
plt.bar(x - 2*bar_width, df['Random Forest MAE'], width=bar_width, label='Random Forest')
plt.bar(x - bar_width, df['LSTM MAE'], width=bar_width, label='LSTM')
plt.bar(x, df['GRU MAE'], width=bar_width, label='GRU')
plt.bar(x + bar_width, df['STGCN MAE'], width=bar_width, label='STGCN')
plt.bar(x + 2*bar_width, df['CNN-GRU-LSTM MAE'], width=bar_width, label='CNN-GRU-LSTM')
# Uncomment when ready
# plt.bar(x + 3*bar_width, df['STGCN-GRU-LSTM MAE'], width=bar_width, label='STGCN-GRU-LSTM')

plt.title('MAE by Model and Forecast Horizon', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
plt.xticks(ticks=x, labels=horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.legend(fontsize=9, ncol=2)  # smaller legend, 2 columns
plt.tight_layout()
plt.savefig('mae_bar_plot.pdf', bbox_inches='tight')
plt.show()
```
