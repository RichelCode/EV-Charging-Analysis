```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Forecast horizons
horizons = [12, 24, 48]

# Model performance
data = {
    'Horizon': horizons,
    'Linear Regression MAE': [696.04, 724.70, 732.64],
    'Random Forest MAE': [586.70, 627.20, 649.33],
    'LSTM MAE': [245.16, 361.16, 421.01],  # 72h still training

    'Linear Regression RMSE': [1027.22, 1062.04, 1069.93],
    'Random Forest RMSE': [866.81, 925.22, 955.91],
    'LSTM RMSE': [388.32, 574.77, 671.78]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Horizon'], df['Linear Regression MAE'], marker='o', label='Linear Regression')
plt.plot(df['Horizon'], df['Random Forest MAE'], marker='s', label='Random Forest')
plt.plot(df['Horizon'], df['LSTM MAE'], marker='^', label='LSTM')

plt.title('MAE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('mae_line_plot.png', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Horizon'], df['Linear Regression RMSE'], marker='o', label='Linear Regression')
plt.plot(df['Horizon'], df['Random Forest RMSE'], marker='s', label='Random Forest')
plt.plot(df['Horizon'], df['LSTM RMSE'], marker='^', label='LSTM')

plt.title('RMSE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('rmse_line_plot.png', dpi=300)
plt.show()
```

```python
width = 0.25  # bar width
x = np.arange(len(horizons))

plt.figure(figsize=(10, 6))
plt.bar(x - width, df['Linear Regression MAE'], width, label='Linear Regression')
plt.bar(x, df['Random Forest MAE'], width, label='Random Forest')
plt.bar(x + width, df['LSTM MAE'], width, label='LSTM')

plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.title('MAE by Model and Forecast Horizon', fontsize=14)
plt.xticks(ticks=x, labels=horizons)
plt.legend()
plt.grid(True, axis='y')
plt.tight_layout()
plt.savefig('mae_bar_plot.png', dpi=300)
plt.show()
```
