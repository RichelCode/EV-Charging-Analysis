```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Updated forecast horizons (now includes 72h)
horizons = [12, 24, 48, 72]

# Updated model performance with 72-hour LSTM results
data = {
    'Horizon': horizons,
    'Linear Regression MAE': [696.04, 724.70, 732.64, 778.93],
    'Random Forest MAE': [586.70, 627.20, 649.33, 685.43],
    'LSTM MAE': [245.16, 361.16, 421.01, 554.02],

    'Linear Regression RMSE': [1027.22, 1062.04, 1069.93, 1120.21],
    'Random Forest RMSE': [866.81, 925.22, 955.91, 995.60],
    'LSTM RMSE': [388.32, 574.77, 671.78, 846.77]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 5))

plt.plot(df['Horizon'], df['Linear Regression MAE'], marker='o', linewidth=2, label='Linear Regression', color='royalblue')
plt.plot(df['Horizon'], df['Random Forest MAE'], marker='s', linewidth=2, label='Random Forest', color='darkorange')
plt.plot(df['Horizon'], df['LSTM MAE'], marker='^', linewidth=2, label='LSTM', color='seagreen')

plt.title('MAE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
plt.xticks(horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(fontsize=11, loc='upper left')
plt.tight_layout()
plt.savefig('mae_line_plot.pdf', bbox_inches='tight')  # High-res for LaTeX
plt.show()
```

```python
plt.figure(figsize=(8, 5))

plt.plot(df['Horizon'], df['Linear Regression RMSE'], marker='o', linewidth=2, label='Linear Regression', color='royalblue')
plt.plot(df['Horizon'], df['Random Forest RMSE'], marker='s', linewidth=2, label='Random Forest', color='darkorange')
plt.plot(df['Horizon'], df['LSTM RMSE'], marker='^', linewidth=2, label='LSTM', color='seagreen')

plt.title('RMSE Across Forecast Horizons', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Root Mean Squared Error (RMSE)', fontsize=12)
plt.xticks(horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.legend(fontsize=11, loc='upper left')
plt.tight_layout()
plt.savefig('rmse_line_plot.pdf', bbox_inches='tight')
plt.show()
```

```python
bar_width = 0.25
x = np.arange(len(horizons))

plt.figure(figsize=(8, 5))
plt.bar(x - bar_width, df['Linear Regression MAE'], width=bar_width, label='Linear Regression', color='royalblue')
plt.bar(x, df['Random Forest MAE'], width=bar_width, label='Random Forest', color='darkorange')
plt.bar(x + bar_width, df['LSTM MAE'], width=bar_width, label='LSTM', color='seagreen')

plt.title('MAE by Model and Forecast Horizon', fontsize=14)
plt.xlabel('Forecast Horizon (hours)', fontsize=12)
plt.ylabel('Mean Absolute Error (MAE)', fontsize=12)
plt.xticks(ticks=x, labels=horizons, fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11)
plt.grid(True, axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig('mae_bar_plot.pdf',bbox_inches='tight')
plt.show()
```
